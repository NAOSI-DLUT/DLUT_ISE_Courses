```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

plt.rcParams['font.family'] = ['SimHei']   # 
plt.rcParams['axes.unicode_minus'] = False # 

x = np.linspace(0, 10, 1000)

# 
plt.figure(figsize=(10,5))

plt.plot(x, np.cos(x), 'b-', label='1')  # 
plt.plot(x, np.cos(x + 1), '--g', label='2') # 
plt.plot(x, np.cos(x + 2), 'r-.', label='3') # 
plt.plot(x, np.cos(x + 3), 'y:', label='4')  # 

plt.title('')  # 

plt.ylabel('y')  # y
plt.xlabel('x')  # x

plt.legend()  #  

#  72 
plt.savefig('', dpi=72)
```
